## Point class and rectangle class

```python
class Point(object):
    """Represents a pint in 2D space.
    
    attributes: x, y
    """

class Rectangle(object):
    """Represents a rectangle.

    attributes: width, height, corner
    """

def print_point(p):
    print('(%g, %g)' % (p.x, p.y))


### Point
# Print the class
print(Point)

# Create an instance
blank = Point()
print(blank)

# Attributes
blank.x = 3.0
blank.y = 4.0

print(blank.y)
x=blank.x
print(blank.x)

print_point(blank)

### Rectangle
box = Rectangle()
box.width = 100.0
box.height = 200.0
box.corner = Point()
box.corner.x = 0.0
box.corner.y = 0.0

### return an instance
def find_center(rect):
    p = Point()
    p.x = rect.corner.x + rect.width/2.0
    p.y = rect.corner.y + rect.height/2.0
    return p


center = find_center(box)
print_point(center)

### Objects are mutable
def grow_rectangle(rect,dwidth,dheight):
    rect.width += dwidth
    rect.height += dheight

print(box.width,box.height)
grow_rectangle(box,50,100)
print(box.width,box.height)

### Copying
p1 = Point()
p1.x = 3.0
p1.y = 4.0

import copy

# pp1 and p1 is the same Point
pp1 = p1
print_point(p1)
print_point(pp1)
print( p1 is pp1 )
print( p1 == pp1 )


# p1 and p2 contains the same data, but not the same Point

p2 = copy.copy(p1)
print_point(p1)
print_point(p2)
print( p1 == p2 )

box2 = copy.copy(box)
print(box2 is box)
print(box2.corner is box.corner)

box3 = copy.deepcopy(box)
print(box3 is box)
print(box3.corner is box.corner)
```

```python
from math import sqrt

class Point(object):
    """Represents a pint in 2D space.
    
    attributes: x, y
    """

class Rectangle(object):
    """Represents a rectangle.

    attributes: width, height, corner
    """

def print_point(p):
    print('(%g, %g)' % (p.x, p.y))

def distance_between_points(p1,p2):
    dx = p1.x - p2.x
    dy = p1.y - p2.y
    return sqrt(dx*dx+dy*dy)

def move_rectangle(p,dx,dy):
    p.x += dx
    p.y += dy

def move_rectangle_c(p,dx,dy):
    new_p = Point()
    new_p.x = p.x+dx
    new_p.y = p.y+dy
    return new_p
    
# create two points and find the distance
A = Point()
A.x=1.0
A.y=2.0
print_point(A)
B = Point()
B.x=1.0+3.0
B.y=2.0+4.0
print_point(B)
print 'distance=', distance_between_points(A,B)

print 'move A'
move_rectangle(A,3,4)
print_point(A)

print 'create B^prime'
Bp = move_rectangle_c(B,3,4)
print_point(B)
print_point(Bp)
```

## Time class

```python
# class definition
class Time(object):
    """Represents the time of day.

    attributes: hour, minute, second
    """

# print_time
def print_time(t):
    print('%.2d:%.2d:%.2d' % (t.hour, t.minute, t.second))
    # print('(%g, %g)' % (p.x, p.y))

time = Time()
time.hour = 11
time.minute = 59
time.second = 30
print 'print_time'
print_time(time)

# time to int and int to time
def time_to_int(time):
    minutes = time.hour * 60 + time.minute
    seconds = minutes * 60 + time.second
    return seconds

def int_to_time(seconds):
    time = Time()
    minutes, time.second = divmod(seconds, 60)
    time.hour, time.minute = divmod(minutes, 60)
    return time

def is_after(t1, t2):
    return time_to_int(t1) > time_to_int(t2)

t1=Time()
t1.hour=3
t1.minute=33
t1.second=30

t2=Time()
t2.hour=2
t2.minute=22
t2.second=22
print 't1:',
print_time(t1)
print 't2:',
print_time(t2)
print is_after(t1,t2)

# prototype of add_tim
# pure function does not modify any argument object
def add_time(t1, t2):
    sum = Time()
    sum.hour = t1.hour + t2.hour
    sum.minute = t1.minute + t2.minute
    sum.second = t1.second + t2.second

    # carry
    if sum.second >= 60:
        sum.second -= 60
        sum.minute +=1

    if sum.minute >= 60:
        sum.minute -= 60
        sum.hour +- 1
    return sum

def add_time_2(t1,t2):
    seconds = time_to_int(t1) + time_to_int(t2)
    return int_to_time(seconds)


start = Time()
start.hour = 9
start.minute = 45
start.second = 0

duration = Time()
duration.hour = 1
duration.minute = 35
duration.second = 0

done = add_time(start, duration)
print 'start time:',
print_time(start)
print 'duration:',
print_time(duration)
print 'end time:',
print_time(done)

# modifiers
def increment(time, seconds):
    time.second += seconds

    if time.second >= 60:
        time.second -= 60
        time.minute +=1

    if time.minute >= 60:
        time.minute -= 60
        time.hour +- 1

def increment_modifier(time, seconds):
    temp = time_to_int(time)
    new_time = int_to_time(temp+seconds)
    time.hour = new_time.hour
    time.minute = new_time.minute
    time.second = new_time.second

def increment_pure(time, seconds):
    temp = time_to_int(time)
    return int_to_time(temp+seconds)

t1=Time()
t1.hour=3
t1.minute=33
t1.second=30
print 't1:',
print_time(t1)
increment_modifier(t1,100)
print 't1+100:',
print_time(t1)
print 't1+200:',
print_time(increment_pure(t1,100))
```
